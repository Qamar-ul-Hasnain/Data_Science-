```python
titanic_data['Age'].fillna(titanic_data['Age'].mean(),inplace=True)
```

```python
titanic_data['Age'].isna().sum()
```

```python
titanic_data.drop('Cabin',axis=1,inplace=True)
```

```python
gender=pd.get_dummies(titanic_data['Sex'],drop_first=True)
titanic_data['Gender']=gender
titanic_data.head()
```

```python
x=titanic_data[['PassengerId','Pclass','Age','SibSp','Parch','Fare','Gender']]
y=titanic_data['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters = {'penalty':['l2','l1','elasticnet'],'C':[1,2,3,4,5,6],'max_iter':[100,200,300,400,500]}
```

```python
data1 = GridSearchCV(classifier,parameters,scoring='accuracy',cv=5)
```

```python
data1.fit(x_train,y_train)
```

```python
print(data1.best_estimator_)
```

```python
print(data1.best_params_)
```

```python
print(data1.best_score_)
```

```python
y_pred = data1.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
acc = accuracy_score(y_pred,y_test)
acc
```

```python
clss = classification_report(y_pred,y_test)
```
